# Graphical View

```python
Graphic1 = pd.read_csv('Merged_data.csv')
```

```python
# Calulating Age
Customer['Age'] = pd.to_datetime('today').year - pd.to_datetime(Customer['Birthday']).dt.year

# Distribution by Age 
sns.histplot(Customer['Age'], bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python

# Preferred Products
preferred_products = Graphic1['Product Name'].value_counts().head(20)

# Donut chat for Preferred Products
plt.figure(figsize=(10, 8))
plt.pie(preferred_products, labels=preferred_products.index, autopct='%1.1f%%', startangle=140)
centre_circle = plt.Circle((0, 0), 0.70, color='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('20 Preferred Products')
plt.tight_layout()
plt.show()
```

```python
# Group by store and sum the 'Order Value'

Graphic1['Order Value'] = Graphic1['Quantity'] * Graphic1['Unit Price USD']

# Group by currency code and sum the 'Order Value'

Sales_by_Currency = Graphic1.groupby('Currency Code')['Order Value'].sum()

# Plotting sales by currency
plt.figure(figsize=(8,6))
plt.pie(Sales_by_Currency, labels=Sales_by_Currency.index, autopct='%1.1f%%', startangle=0)
plt.title('Sales Performance by Currency ')
plt.show()
```

```python
# Calculate Order Value (Total Revenue) for each product
Graphic1['Order Value'] = Graphic1['Quantity'] * Graphic1['Unit Price USD']

# Calculate profit for each product
Graphic1['Profit'] = Graphic1['Unit Price USD'] - Graphic1['Unit Cost USD']

# Calculate total profit for each product based on quantity sold
Graphic1['Total Profit'] = Graphic1['Profit'] * Graphic1['Quantity']

# Calculate profit margin for each product
Graphic1['Profit Margin'] = (Graphic1['Profit'] / Graphic1['Unit Price USD']) * 100  # Profit Margin as a percentage of selling price

# Group by 'Product Name' and calculate total revenue, total profit, and average profit margin
profitability = Graphic1.groupby('Product Name').agg(
    Total_Revenue=('Order Value', 'sum'),
    Total_Profit=('Total Profit', 'sum'),
    Avg_Profit_Margin=('Profit Margin', 'mean')
).reset_index()

# profit margin
top_profitable_products = profitability.sort_values(by='Avg_Profit_Margin')

plt.figure(figsize=(10, 6))
sns.violinplot(x=profitability['Avg_Profit_Margin'], color='green')
plt.title('Violin Plot of Average Profit Margin Across Products')
plt.xlabel('Profit Margin (%)')
plt.show()
```

```python
# Calculate total quantity for each Category
Category_performance = Graphic1.groupby(['Category']).agg(
    Total_Quantity=('Quantity', 'sum')
).reset_index()

# Sort subcategories by total quantity
Category_performance = Category_performance.sort_values(by='Total_Quantity', ascending=False)

# Plotting a simple bar chart
plt.figure(figsize=(15, 10))
sns.barplot(data=Category_performance, x='Category', y='Total_Quantity')
plt.title('Total Quantity Sold by Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
